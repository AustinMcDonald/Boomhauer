```python
import math

def smear_evt_energy(mcE       : float,
                     sigma_Qbb : float,
					 Qbb       : float
					) -> float:
	"""
	It smears and returns the montecarlo energy of the event (mcE) according to:
	the sigma at Qbb (sigma_Qbb) in absolute values (keV).
	"""
	sigma_E = sigma_Qbb * math.sqrt(mcE/Qbb)
	smE = np.random.normal(mcE, sigma_E)
	return smE
```

```python
def MC_to_electrons(MC_HIT,Space_sigma):
    #Space_sigma = 0.1 # 0.1 should be 100 microns
    #MC_HIT = Events[0][2]
    mu = int(MC_HIT.E/Ionization) # picks the number of electrons
    #print(mu)
    si = np.sqrt(mu) # decides there error
    New_mu = int(np.random.normal(mu, si, 1))# gererates a gaussian for that with the number of electrons
    #print(New_mu)
    X_mu =MC_HIT.pos[0] ; Y_mu = MC_HIT.pos[1] ; Z_mu = MC_HIT.pos[2]
    #Xs = [] ;Ys=[];Zs=[]
    New_pos = []
    # Wrong but is a gaussian distrbution
    #for po in range(New_mu):
    #    xnew = np.random.normal(X_mu, Space_sigma, 1)[0]
    #    ynew = np.random.normal(Y_mu, Space_sigma, 1)[0]
    #    znew = np.random.normal(Z_mu, Space_sigma, 1)[0]
    #    New_pos.append([xnew,ynew,znew])
    for po in range(New_mu):
        xnew = np.random.uniform(X_mu-Space_sigma, X_mu+Space_sigma, 1)[0]
        ynew = np.random.uniform(Y_mu-Space_sigma, Y_mu+Space_sigma, 1)[0]
        znew = np.random.uniform(Z_mu-Space_sigma, Z_mu+Space_sigma, 1)[0]
        New_pos.append([xnew,ynew,znew])
    
    return New_pos
```

```python
def Event_to_electrons(MC_HITs,Space_sigma):
    Event_electrons = []
    for qw in range(0,len(MC_HITs)):
        hold = MC_to_electrons(MC_HITs[qw],Space_sigma)
        Event_electrons+=hold
    Event_electrons = np.array(Event_electrons)
    return Event_electrons
```

```python
def Diffuser(Event_electrons,Gas_props):
    Dif_Tran_star  = Gas_props[0]#*1e-2
    Dif_Long_star  = Gas_props[1]#*1e-2
    Drift_Vel      = Gas_props[2]  #units mm/mus
    Pressure       = Gas_props[3]  #units of bar
    Life_Time      = Gas_props[4]  #units mus
    
    Diffused = []
    for ty in range(0,len(Event_electrons)):
        X = Event_electrons[ty][0]
        Y = Event_electrons[ty][1]
        Z = Event_electrons[ty][2]

        Drift_distance = (5000 + Z)#/1000
        DT = Dif_Tran_star*np.sqrt(Drift_distance*1e-1/Pressure)*1e-3
        DL = Dif_Long_star*np.sqrt(Drift_distance*1e-1/Pressure)*1e-3
        #DT = Dif_Tran_star*np.sqrt(Drift_distance*1e3)/np.sqrt(Pressure)*1e-5
        #DL = Dif_Long_star*np.sqrt(Drift_distance*1e3)/np.sqrt(Pressure)*1e-5
        #DT = DT/1000
        #DL = DL/1000

        #Drift_time     = Drift_distance/(Drift_Vel/10)
        Drift_time     = Drift_distance/(Drift_Vel)
        Life_time_num = np.random.exponential(Drift_time/Life_Time,1)[0]
        if Life_time_num <= Drift_time/Life_Time:

            #xnew = np.random.uniform(X-DT/2, X+DT/2, 1)[0]
            #ynew = np.random.uniform(Y-DT/2, Y+DT/2, 1)[0]
            #znew = np.random.uniform(Z-DL/2, Z+DL/2, 1)[0]
            xnew = np.random.normal(X, DT, 1)[0]
            ynew = np.random.normal(Y, DT, 1)[0]
            znew = np.random.normal(Z, DL, 1)[0]
            Diffused.append([xnew,ynew,znew])
    Diffused = np.array(Diffused)
    return Diffused
```

```python
def gaussian(x, height, center, width):
    return height*np.exp(-(x - center)**2/(2*width**2))
def one_gaussians(x, h1, c1, w1):
    return (gaussian(x, h1, c1, w1))
def gauss_fit_vals(X,Y):
    minn = max(Y)
    MeanLoc = np.where(Y==minn)[0][0]
    Mean = X[MeanLoc]
    vals = [minn,Mean,20]
    return vals
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
from scipy import optimize
```

```python
def chi(Ztrue,Zreco,bins):
    xx = Ztrue
    yy = Zreco
    bb = 0
    ch = []
    for x in range(0,len(bins)):
        edg = bins[x]
        #nn = xx[(xx>bb) & (xx<edg)]
        nn = [(xx>bb) & (xx<edg)]
        #nn = [(yy>bb) & (yy<edg)]
        ll = len(xx[nn])
        if ll ==0:
            ll=1
        Delta_R = np.sqrt(sum((xx[nn]-yy[nn])**2)/ll)
        ch.append(Delta_R)
        bb = edg
    return ch
```
